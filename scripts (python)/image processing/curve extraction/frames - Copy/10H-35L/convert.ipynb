```python
import cv2
import glob
png_file_paths  = glob.glob('*.png') #glob.glob('*.jpg')
#print(images)
for i, png_file_path in enumerate(png_file_paths):
    jpg_file_path = png_file_path[:-4] + ".jpg"
    img = cv2.imread(png_file_path)
    # Save .jpg image
    cv2.imwrite(jpg_file_path, img, [int(cv2.IMWRITE_JPEG_QUALITY), 100])
```
